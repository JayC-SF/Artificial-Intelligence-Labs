```python
import numpy as np
```

```python

def manhattan(points, centroids):
    distances = np.zeros((len(points), len(centroids)), dtype=float)
    for i, p in enumerate(points):
        px, py = p
        for j, c in enumerate(centroids):
            cx, cy = c
            distances[i, j] = abs(px-cx)+abs(py-cy)
            # print(f"P{i+1} C{j+1}", distances[i,j])
    # calculate new points
    points_c_index = np.argmin(distances, axis=1)
    new_centroids = np.zeros(centroids.shape)
    for j in range(len(centroids)):
        total = np.array([0 , 0])
        count = 0.0
        for i, p in enumerate(points_c_index):
            if p == j:
                total += points[i]
                count += 1
        total = total/count
        new_centroids[j] = total
    return distances, new_centroids
```

```python
points = np.array([
    [1, 2],
    [2, 7],
    [3, 3],
    [6, 4],
    [7, 8],
    [8, 2],
    [9, 1],
])
centroids = np.array([
    [2, 7],
    [6, 4],
    [8, 2],
])
```

```python
iteration = 0
new_centroids = centroids
while True:
    iteration += 1
    print(f"Iteration {iteration}")
    prev_centroids = new_centroids
    points_c_index, new_centroids = manhattan(points, prev_centroids)
    print(points_c_index)
    print(new_centroids)
    if np.all(prev_centroids == new_centroids):
        break
```
