```python
import torch
import torch.nn as nn
```

```python
# Input: 32x32x3
x = torch.randn(1, 3, 32, 32)  # (batch, channels, height, width)


# Layer 1: Conv2d 16 filters, 3x3, stride 1, pad 1
conv1 = nn.Conv2d(in_channels=3, out_channels=16, kernel_size=3, stride=1, padding=1)

x = conv1(x)
print("After Conv2D1:", x.shape)

# ReLU doesn't change shape
# MaxPool 2x2, stride 2
pool1 = nn.MaxPool2d(kernel_size=2, stride=2)
x = pool1(x)
print("After MaxPool1:", x.shape)

conv2 = nn.Conv2d(in_channels=16, out_channels=32, kernel_size=3, stride=1, padding=0)
x = conv2(x)
print("After Conv2D2:", x.shape)

pool2 = nn.MaxPool2d(kernel_size=2, stride=2)
x = pool2(x)
print("After MaxPool2:", x.shape)

conv3 = nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3, stride=1, padding=0)
x = conv3(x)
print("After Conv2D3", x.shape)

flatten = nn.Flatten()
x = flatten(x)
print("After flatten:", x.shape)

linear1 = nn.Linear(in_features=1600, out_features=128)
x = linear1(x)
print("After linear1:", x.shape)

linear1 = nn.Linear(in_features=128, out_features=10)
x = linear1(x)
print("After linear2:", x.shape)


# Continue adding layers...
```
