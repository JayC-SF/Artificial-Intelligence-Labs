```python
import torch
import torch.nn as nn
```

```python
x = torch.randn(1,3,128,128)
print("Input shape:", x.shape)
conv1 = nn.Conv2d(in_channels=3, out_channels=16, kernel_size=5, padding=2, stride=2)
x = conv1(x)
print("After conv1:", x.shape) # shape should be 64x64

pool1 = nn.MaxPool2d(kernel_size=2, stride=2)
x = pool1(x)
print("After pool1:", x.shape) # should be 32x32

conv2 = nn.Conv2d(in_channels=16, out_channels=32, kernel_size=3, padding=1, stride=2)
x = conv2(x)
print("After conv2:", x.shape) # should be 16x16

pool1 = nn.MaxPool2d(kernel_size=2, stride=2)
x = pool1(x)
print("After pool1:", x.shape) # should be 8x8

conv3 = nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3, padding=1, stride=2)
x = conv3(x)
print("After conv3:", x.shape) # should be 4x4x64

flatten = nn.Flatten()
x = flatten(x)
print("After flatten:", x.shape)

linear = nn.Linear(in_features=1024, out_features=8)
x = linear(x)
print("After linear:", x.shape)
```
